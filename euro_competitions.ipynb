```python
# import all necessary packages for webscraping a list of Bundesliga first devision teams

import time
import requests
from bs4 import BeautifulSoup
import re # support regular expressions
import pandas as pd
import numpy as np
```

```python
# define target page and assign User Agent to mask browser use

url = 'https://www.transfermarkt.de/wettbewerbe/europa?ajax=yw1&page='

headers = {'User-Agent': 
           'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}
```

```python
# define a function to adjust the needed url for every single page of content and returns every pages content as a combined list

def get_all_pages(str_url):
    pages = list(range(1, 20))
    comp_names_list = []
    # return [(url + str(page)) for page in pages]
    for page in pages:
        url = 'https://www.transfermarkt.de/wettbewerbe/europa?ajax=yw1&page=' + str(page)
        response = requests.get(url, headers=headers)
        html_soup = BeautifulSoup(response.content, 'html.parser')
        comp_table = html_soup.find(id='yw1')
        comp_elements = comp_table.find_all('td', {'class': 'hauptlink'})
        comp_names = [comp.text.replace(u'\n', u'').strip() for comp in comp_elements][1::2]
        comp_names_list.extend(comp_names)    

    return comp_names_list

comp_names = get_all_pages(url)
comp_names
```

```python
df = pd.DataFrame(pd.Series(comp_names), columns= ['competition'])
df
```
